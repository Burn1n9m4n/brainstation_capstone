```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns',None)
```

```python
vectorized_df = pd.read_parquet('../data/20230809_complete_vectorized_data.parquet')
```

```python
vectorized_df.shape
```

```python
vectorized_df.loc[:,['24','25','26','27']].mean()
```

```python
vectorized_df.shape
```

```python
x = pairwise.cosine_similarity(vectorized_df, dense_output=True)
```

```python
sort_x = np.argsort(x)[:,::-1]
```

```python
sort_x.shape
```

```python
sort_x
```

```python
vec = list(sort_x[0,0:6])
vec
```

```python
tracks = list(vectorized_df.iloc[vec,:].index)
tracks
```

```python
kaggle = pd.read_csv('../data/SpotifyFeatures.csv')
```

```python
kaggle['track_id'] = kaggle['track_id'].astype("category")
kaggle['track_id'] = kaggle.track_id.cat.set_categories(tracks) # sets this as a category and provides a hierarchy
```

```python
track_recommendations = kaggle.copy(deep=True).sort_values(["track_id"]).head(6) # sorts heirarchy
track_recommendations
```
