### Priliminary EDA

```python
df.artist_name.value_counts(normalize=True).reset_index().head()
```

```python
pareto = df.artist_name.value_counts(normalize=True).cumsum().reset_index()
pareto["proportion"] = pareto.proportion * 100
pareto.loc[pareto.proportion <= 80].shape[0] / pareto.shape[0] * 100
```

Roughly 23% of the artists within the data set represent 80% of the data.
